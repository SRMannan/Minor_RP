```python
import spacy
from spacy.tokens import DocBin
from spacy.training.example import Example
from spacy.training import Example
import json

# Load the transformer-based pipeline for Hindi (MuRIL)
nlp = spacy.blank("hi")  # Generic spaCy pipeline for any language, replace with "hi" if it's supported

# Add the NER pipeline
ner = nlp.add_pipe("ner")

# Load JSON data
with open("/content/combined_annotations234.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Add labels
for label in data["classes"]:
    ner.add_label(label)

# Convert annotations to spaCy format
examples = []
for text, annotation in data["annotations"]:
    doc = nlp.make_doc(text)
    example = Example.from_dict(doc, annotation)
    examples.append(example)

# Train the NER model
optimizer = nlp.initialize()

# Training loop
for epoch in range(100):  # Set a number of epochs based on your needs
    losses = {}
    nlp.update(examples, sgd=optimizer, losses=losses)
    print(f"Epoch {epoch+1} - Loss: {losses['ner']}")

# Save the trained model
nlp.to_disk("hindi_ner_model")
```
